# Data Cleaning

```python
col=[]
vc=[]
per=[]
for i in df:
    null=df[i].isnull().sum()    #null value count
    if null>0:
        persentage=(null/df.shape[0])*100
        col.append(i)
        per.append(persentage)
        vc.append(null)
nvp={'column name': col,'null value':vc,'percentage':per }    #nvp is null alue percentage value count
pnv=pd.DataFrame(nvp)   #pnv is persentage value count
pnv
```

```python
pnv2=pnv[pnv['percentage']>25]
```

```python
ctr = []
for i in pnv2.iloc[:,0]:
    ctr.append(i)
```

```python
for i in ctr:
    df.drop(columns=i,inplace=True)
```

```python
df
```

```python
pnv3=pnv[pnv['percentage']<25]
```

# which univercity have most admission

```python
yoa=df.groupby('university_name')['university_name'].size().rename('coa').reset_index()
yoa.sort_values(by='coa',ascending=False, inplace=True)
```

```python
plt.figure(figsize=(8,7))
sns.barplot(y=yoa['university_name'].head(10),x=yoa['coa'].head(10),palette='Greens',hue=yoa['university_name'].head(10))
plt.title("TOP10 Universities (Admission's Count)")
plt.ylabel('University Names',fontdict={'weight':'normal','size':'15'})
plt.xlabel('Count of Admissions',fontdict={'weight':'normal','size':'15'})
plt.show()
plt.figure(figsize=(8,7))
sns.barplot(y=yoa['university_name'].tail(10),x=yoa['coa'].tail(10),palette='Reds',hue=yoa['university_name'].tail(10))
plt.title("BOTTOM10 Universities (Admission's Count)")
plt.ylabel('University Names',fontdict={'weight':'normal','size':'15'})
plt.xlabel('Count of Admissions',fontdict={'weight':'normal','size':'15'})
plt.show()
```
