# Data Cleaning

```python
col=[]
vc=[]
per=[]
for i in df:
    null=df[i].isnull().sum()    #null value count
    if null>0:
        persentage=(null/df.shape[0])*100
        col.append(i)
        per.append(persentage)
        vc.append(null)
nvp={'column name': col,'null value':vc,'percentage':per }    #nvp is null alue percentage value count
pnv=pd.DataFrame(nvp)   #pnv is persentage value count
pnv
```

```python
pnv2=pnv[pnv['percentage']>25]
```

```python
ctr = []
for i in pnv2.iloc[:,0]:
    ctr.append(i)
```

```python
for i in ctr:
    df.drop(columns=i,inplace=True)
```
